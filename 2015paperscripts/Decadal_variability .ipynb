```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
from numba import autojit
from scipy.interpolate import interp1d
import glob
sys.path.append('/noc/users/hb1g13/Python/python_functions/')
import useful as hb
import SG
```

```python
Year='0-300'
tau='3daynokpp'
x='/hpcdata/scratch/hb1g13/NchannelFlat'
```

```python
gridfilepath=x+'/'+tau
file2=netcdf.netcdf_file(gridfilepath+'/grid.nc','r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
Z=file2.variables['Z']
Z=Z[:]*1
Y=file2.variables['Yp1']
Y=Y[:]*1
Xp=file2.variables['Xp1']
Xp=Xp[:]*1
Yc=file2.variables['Y']
Yc=Yc[:]*1
```

```python
fig=plt.figure(figsize=(25.5,7.5))
fname=x+'/'+tau+'/'+Year+'psi.nc'
file2read = netcdf.NetCDFFile(fname,'r')
lvrho=file2read.variables["LaVH1TH"]
lvrho=lvrho[:]
time=file2read.variables['T']
ti=time[:]
dx=Y[1]-Y[0] # Find Resolution
VT=np.sum(lvrho*dx,axis=3) #integrate Vdx along x
VTfdz=np.cumsum(VT[:,::-1,:],axis=1) #sum up the water column
psi=VTfdz[:,::-1,:]/10**6 #Time average and put into Sv and put back in right order
a=list(fname)
a[-6]='A'
a[-5]='l'
a[-4]='l'
A=''.join(a)
file2 = netcdf.NetCDFFile(A,'r')
U=file2.variables['UVEL']
U=U[:]*1
V=file2.variables['VVEL']
V=V[:]*1
Vzone=np.nansum(V*dx,axis = 3)
dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
# No more super slow forloop!           
psi2=np.apply_along_axis(np.multiply,1,Vzone,dz)
psi3=np.cumsum(-psi2[:,::-1,:],axis=1)
npad = ((0,0),(0,1), (0,0))
psi4 = np.pad(psi3, pad_width=npad, mode='constant', constant_values=0)
Psi=np.max(np.max(psi,axis=1),axis=1)
Psi3=np.min(np.min(psi,axis=1),axis=1)
Psi2=np.max(np.max(psi4/10**6,axis=1),axis=1)
ACC=np.zeros((300,len(Z),len(Yc),len(Xp)))
for t in range(len(U[:,1,1,1])):
    for ix in range(len(Yc)):
        for j in range(len(Xp)):
            for k in range(0,len(Z)-1):
                ACC[t,k,ix,j] = dz[k]*U[t,k,ix,j]
ACCf=np.cumsum(ACC[:,::-1,:,:],axis=1)
ACC=np.mean(np.sum(ACCf[:,::-1,100:300,:],axis=2),axis=2)
ACC=np.max(ACC,axis=1)
ax = fig.add_subplot(1, 4, 1)
p=ax.plot(Psi2) #Use b2r colourmap
ax.set_title(tau+' MOC',fontsize=20)
ax.set_xlabel('Time (yrs)',fontsize=20)
ax.set_ylabel('Transport (Sv)',fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = fig.add_subplot(1, 4, 2)
p=ax.plot(Psi) #User b2r colourmap
ax.set_title(tau+' ROC',fontsize=20)
ax.set_xlabel('Time (yrs)',fontsize=20)
ax.set_ylabel('Transport (Sv)',fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = fig.add_subplot(1, 4, 3)
p=ax.plot(Psi3) #Use b2r colourmap
ax.set_title(tau+' ROC (neg)',fontsize=20)
ax.set_xlabel('Time (yrs)',fontsize=20)
ax.set_ylabel('Transport (yrs)',fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = fig.add_subplot(1, 4, 4)
p=ax.plot(ACC) #Use b2r colourmap
ax.set_title(tau+' ACC (neg)',fontsize=20)
ax.set_xlabel('Time (yrs)',fontsize=20)
ax.set_ylabel('Transport (Sv)',fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
plt.tight_layout()
```
